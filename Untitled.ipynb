```python
# =========================================
# Objective 2: Effect of NaOH Treatment on Compressive & Flexural Strength
# =========================================
# pip install pandas numpy scipy seaborn matplotlib openpyxl

import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# 1) Load dataset
# -------------------------
DATA_PATH = "dataset.xlsx"   # change filename if needed
df = pd.read_excel(DATA_PATH)

# Clean column names (remove spaces, unify case)
df.columns = df.columns.str.strip()

# Convert numeric columns
df['FORCE (kN)'] = pd.to_numeric(df['FORCE (kN)'], errors='coerce')
df['Strength (MPa)'] = pd.to_numeric(df['Strength (MPa)'], errors='coerce')

# Preview dataset
print("\n=== Dataset Preview ===")
print(df.head())
print("\nUnique Batch Types:", df['Batch type'].unique())

# -------------------------
# 2) Split into compressive and flexural sets
# -------------------------
compressive = df[df['Batch type'].str.contains("COMPRESSIVE", case=False)].copy()
flexural = df[df['Batch type'].str.contains("FLEXURAL", case=False)].copy()

print(f"\nCompressive samples: {len(compressive)}, Flexural samples: {len(flexural)}")

# -------------------------
# 3) Descriptive Statistics
# -------------------------
def descriptive_stats(data, test_name):
    print(f"\n=== {test_name} Strength Summary ===")
    summary = data.groupby('Batch type')['Strength (MPa)'].agg(['count','mean','std','min','max']).round(3)
    print(summary)
    return summary

comp_summary = descriptive_stats(compressive, "Compressive")
flex_summary = descriptive_stats(flexural, "Flexural")

# -------------------------
# 4) Check for Normality (Shapiro–Wilk)
# -------------------------
def check_normality(data, test_name):
    print(f"\n=== {test_name} Normality Test (Shapiro–Wilk) ===")
    for batch in data['Batch type'].unique():
        values = data.loc[data['Batch type']==batch, 'Strength (MPa)']
        if len(values) >= 3:
            stat, p = stats.shapiro(values)
            print(f"{batch}: W={stat:.3f}, p={p:.4f} → {'Normal' if p>=0.05 else 'Not normal'}")
        else:
            print(f"{batch}: Not enough data for test")

check_normality(compressive, "Compressive")
check_normality(flexural, "Flexural")

# -------------------------
# 5) Compare Batches (T-test or Mann–Whitney)
# -------------------------
def compare_batches(data, test_name):
    batches = data['Batch type'].unique()
    if len(batches) != 2:
        print(f"{test_name}: Need exactly 2 batches for comparison.")
        return

    g1 = data.loc[data['Batch type']==batches[0], 'Strength (MPa)'].dropna()
    g2 = data.loc[data['Batch type']==batches[1], 'Strength (MPa)'].dropna()
    
    # Normality check
    normal1 = stats.shapiro(g1)[1] >= 0.05 if len(g1)>=3 else False
    normal2 = stats.shapiro(g2)[1] >= 0.05 if len(g2)>=3 else False

    print(f"\n=== {test_name} Statistical Comparison ===")
    print(f"{batches[0]} mean = {g1.mean():.3f} MPa, {batches[1]} mean = {g2.mean():.3f} MPa")

    if normal1 and normal2:
        # Welch’s t-test
        t, p = stats.ttest_ind(g1, g2, equal_var=False)
        test_used = "Welch's t-test"
    else:
        # Mann–Whitney U
        t, p = stats.mannwhitneyu(g1, g2, alternative='two-sided')
        test_used = "Mann–Whitney U"

    print(f"{test_used}: statistic={t:.3f}, p={p:.4f}")
    if p < 0.05:
        print("✅ Significant difference (p < 0.05)")
    else:
        print("❌ No significant difference (p ≥ 0.05)")

compare_batches(compressive, "Compressive Strength")
compare_batches(flexural, "Flexural Strength")

# -------------------------
# 6) Visualizations
# -------------------------
sns.set(style="whitegrid")

# Boxplots for compressive
plt.figure(figsize=(8,5))
sns.boxplot(x='Batch type', y='Strength (MPa)', data=compressive, palette='Blues', showmeans=True)
plt.title("Compressive Strength by Batch")
plt.ylabel("Strength (MPa)")
plt.xlabel("")
plt.tight_layout()
plt.savefig("compressive_strength_boxplot.png", dpi=300)
plt.show()

# Boxplots for flexural
plt.figure(figsize=(8,5))
sns.boxplot(x='Batch type', y='Strength (MPa)', data=flexural, palette='Greens', showmeans=True)
plt.title("Flexural Strength by Batch")
plt.ylabel("Strength (MPa)")
plt.xlabel("")
plt.tight_layout()
plt.savefig("flexural_strength_boxplot.png", dpi=300)
plt.show()

# -------------------------
# 7) Summary Output
# -------------------------
comp_summary.to_csv("compressive_summary.csv")
flex_summary.to_csv("flexural_summary.csv")
print("\n✅ Analysis complete. CSV and PNG files saved:")
print(" - compressive_summary.csv")
print(" - flexural_summary.csv")
print(" - compressive_strength_boxplot.png")
print(" - flexural_strength_boxplot.png")
```

```python
# =========================================
# Load–Strength Modeling and Consistency Analysis
# =========================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# -------------------------
# 1) Load data
# -------------------------
DATA_PATH = "dataset.xlsx"
df = pd.read_excel(DATA_PATH)

# Clean column names
df.columns = df.columns.str.strip()

# Convert numeric
df['FORCE (kN)'] = pd.to_numeric(df['FORCE (kN)'], errors='coerce')
df['Strength (MPa)'] = pd.to_numeric(df['Strength (MPa)'], errors='coerce')

# -------------------------
# 2) Linear regression per batch
# -------------------------
print("\n=== Linear Regression: Load–Strength Relationship per Batch ===")

results = []

for batch in df['Batch type'].unique():
    sub = df[df['Batch type'] == batch].dropna(subset=['FORCE (kN)', 'Strength (MPa)'])
    if len(sub) < 3:
        continue
    
    X = sub[['FORCE (kN)']].values
    y = sub['Strength (MPa)'].values

    model = LinearRegression().fit(X, y)
    slope = model.coef_[0]
    intercept = model.intercept_
    r2 = r2_score(y, model.predict(X))

    results.append({
        'Batch type': batch,
        'Slope (MPa per kN)': slope,
        'Intercept': intercept,
        'R²': r2,
        'n': len(sub)
    })

    print(f"{batch}: Strength = {intercept:.3f} + {slope:.3f}×Force, R² = {r2:.3f}")

# Convert to DataFrame
reg_summary = pd.DataFrame(results)
print("\n=== Regression Summary ===")
print(reg_summary.round(4))

# -------------------------
# 3) Visualize relationships
# -------------------------
sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(8,6))
sns.lmplot(data=df, x='FORCE (kN)', y='Strength (MPa)', hue='Batch type', ci=None, height=6, aspect=1.2)
plt.title("Load–Strength Relationship by Batch Type")
plt.xlabel("Force (kN)")
plt.ylabel("Strength (MPa)")
plt.tight_layout()
plt.savefig("load_strength_relationship.png", dpi=300)
plt.show()

# -------------------------
# 4) Coefficient of Variation (CV%) Analysis
# -------------------------
print("\n=== Coefficient of Variation (CV%) per Batch ===")

cv_summary = df.groupby('Batch type')['Strength (MPa)'].agg(['count', 'mean', 'std']).round(3)
cv_summary['CV%'] = (cv_summary['std'] / cv_summary['mean']) * 100
print(cv_summary)

# -------------------------
# 5) Aggregate by Test Type
# -------------------------
df['Test Type'] = df['Batch type'].apply(lambda x: 'Compressive' if 'COMPRESSIVE' in x.upper() else 'Flexural')
cv_test = df.groupby('Test Type')['Strength (MPa)'].agg(['count', 'mean', 'std']).round(3)
cv_test['CV%'] = (cv_test['std'] / cv_test['mean']) * 100
print("\n=== Coefficient of Variation by Test Type ===")
print(cv_test)

# -------------------------
# 6) Save outputs
# -------------------------
reg_summary.to_csv("regression_summary.csv", index=False)
cv_summary.to_csv("cv_summary_batches.csv")
cv_test.to_csv("cv_summary_tests.csv")

print("\n✅ Analysis complete. Files saved:")
print(" - regression_summary.csv")
print(" - cv_summary_batches.csv")
print(" - cv_summary_tests.csv")
print(" - load_strength_relationship.png")
```

```python
# =========================================
# RSM Analysis (CCD) — Find Optimal Fiber % and Temperature
# =========================================
# Requirements:
# pip install pandas numpy scipy matplotlib seaborn statsmodels openpyxl scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from sklearn.preprocessing import MinMaxScaler
import os

# -------------------------
# USER CONFIG
# -------------------------
DATA_PATH = "dataset2.xlsx"   # <-- change to your Excel filename
SHEET_NAME = 0                       # or sheet name
SAVE_FIGS = True
OUT_DIR = "rsm_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# -------------------------
# 1) LOAD & SANITY CHECK
# -------------------------
df = pd.read_excel(DATA_PATH, sheet_name=SHEET_NAME)
df.columns = df.columns.str.strip()

required_cols = ['ID', 'FORCE (kN)', 'Strength (MPa)', 'BFC (%)', 'Temperature (°C)', 'Duration (min)', 'Batch type']
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns in {DATA_PATH}: {missing}")

# Ensure numeric
for c in ['FORCE (kN)', 'Strength (MPa)', 'BFC (%)', 'Temperature (°C)', 'Duration (min)']:
    df[c] = pd.to_numeric(df[c], errors='coerce')

print("Loaded dataset with rows:", len(df))
print("Batch types:", df['Batch type'].unique())

# -------------------------
# 2) Split by Test Type
# -------------------------
df['TestType'] = df['Batch type'].apply(lambda x: 'Compressive' if 'COMPRESS' in x.upper() else ('Flexural' if 'FLEX' in x.upper() else 'Other'))
print(df['TestType'].value_counts())

comp_df = df[df['TestType']=='Compressive'].dropna(subset=['BFC (%)','Temperature (°C)','Strength (MPa)']).copy()
flex_df = df[df['TestType']=='Flexural'].dropna(subset=['BFC (%)','Temperature (°C)','Strength (MPa)']).copy()

# Quick head
print("\nCompressive rows:", len(comp_df), " | Flexural rows:", len(flex_df))

# -------------------------
# Helper: build quadratic design matrix
# -------------------------
def build_quadratic_X(df_in, factor1='BFC (%)', factor2='Temperature (°C)'):
    X = pd.DataFrame({
        'const': 1.0,
        factor1: df_in[factor1],
        factor2: df_in[factor2],
        f'{factor1}*{factor2}': df_in[factor1]*df_in[factor2],
        f'{factor1}^2': df_in[factor1]**2,
        f'{factor2}^2': df_in[factor2]**2
    })
    return X

# -------------------------
# 3) Fit quadratic RSM (OLS) and show ANOVA + summary
# -------------------------
def fit_rsm(df_in, response_col='Strength (MPa)', name='Compressive'):
    print(f"\n--- RSM for {name} (n={len(df_in)}) ---")
    X = build_quadratic_X(df_in)
    y = df_in[response_col]
    model = sm.OLS(y, X).fit()
    print(model.summary())
    try:
        anova_res = anova_lm(model, typ=2)
        print("\nANOVA (Type II):")
        print(anova_res)
    except Exception as e:
        print("ANOVA error:", e)
    return model, X, y

comp_model, comp_X, comp_y = fit_rsm(comp_df, name='Compressive')
flex_model, flex_X, flex_y = fit_rsm(flex_df, name='Flexural')

# -------------------------
# 4) Basic diagnostics plots
# -------------------------
def diagnostics_plots(model, X, y, tag):
    resid = model.resid
    fitted = model.fittedvalues

    plt.figure(figsize=(6,4))
    plt.scatter(fitted, resid)
    plt.axhline(0, color='k', linestyle='--', linewidth=0.6)
    plt.xlabel('Fitted')
    plt.ylabel('Residuals')
    plt.title(f'{tag}: Residuals vs Fitted')
    if SAVE_FIGS: plt.savefig(f"{OUT_DIR}/{tag}_resid_vs_fitted.png", dpi=300)
    plt.show()

    # QQ plot
    sm.qqplot(resid, line='45', fit=True)
    plt.title(f'{tag}: QQ plot of residuals')
    if SAVE_FIGS: plt.savefig(f"{OUT_DIR}/{tag}_qq.png", dpi=300)
    plt.show()

diagnostics_plots(comp_model, comp_X, comp_y, 'Compressive')
diagnostics_plots(flex_model, flex_X, flex_y, 'Flexural')

# -------------------------
# 5) Create prediction grid & find maxima for each response
# -------------------------
# grid ranges: expand a bit around observed data
def prediction_grid_and_optimize(model, df_in, f1='BFC (%)', f2='Temperature (°C)', grid_n=101):
    f1_min, f1_max = df_in[f1].min(), df_in[f1].max()
    f2_min, f2_max = df_in[f2].min(), df_in[f2].max()
    # extend slightly
    pad_f1 = (f1_max - f1_min) * 0.1 if f1_max>f1_min else 0.1
    pad_f2 = (f2_max - f2_min) * 0.1 if f2_max>f2_min else 1.0
    f1_vals = np.linspace(max(0, f1_min-pad_f1), f1_max+pad_f1, grid_n)
    f2_vals = np.linspace(max(0, f2_min-pad_f2), f2_max+pad_f2, grid_n)
    F1, F2 = np.meshgrid(f1_vals, f2_vals)
    # build design terms
    DF = pd.DataFrame({
        'const': 1.0,
        f1: F1.ravel(),
        f2: F2.ravel(),
        f'{f1}*{f2}': (F1*F2).ravel(),
        f'{f1}^2': (F1**2).ravel(),
        f'{f2}^2': (F2**2).ravel()
    })
    preds = model.predict(DF)
    preds_mat = preds.values.reshape(F1.shape)
    # find maximum
    idx = np.unravel_index(np.nanargmax(preds_mat), preds_mat.shape)
    opt_f1 = F1[idx]
    opt_f2 = F2[idx]
    opt_pred = preds_mat[idx]
    return (F1, F2, preds_mat, opt_f1, opt_f2, opt_pred)

comp_F1, comp_F2, comp_pred, comp_opt_f1, comp_opt_f2, comp_opt_pred = prediction_grid_and_optimize(comp_model, comp_df)
flex_F1, flex_F2, flex_pred, flex_opt_f1, flex_opt_f2, flex_opt_pred = prediction_grid_and_optimize(flex_model, flex_df)

print(f"\nCompressive optimum (grid search): BFC = {comp_opt_f1:.4g} %, Temp = {comp_opt_f2:.3f} °C -> Pred Strength = {comp_opt_pred:.4f} MPa")
print(f"Flexural optimum (grid search): BFC = {flex_opt_f1:.4g} %, Temp = {flex_opt_f2:.3f} °C -> Pred Strength = {flex_opt_pred:.4f} MPa")

# -------------------------
# 6) Plot predicted surfaces (contour + surface-like heatmap)
# -------------------------
def contour_and_surface(F1, F2, pred_mat, title, prefix):
    plt.figure(figsize=(8,6))
    cs = plt.contourf(F1, F2, pred_mat, cmap='viridis', levels=30)
    plt.colorbar(cs, label='Predicted Strength (MPa)')
    plt.scatter(df['BFC (%)'], df['Temperature (°C)'], c='white', s=40, edgecolor='k', alpha=0.8, label='Experimental runs')
    plt.xlabel('BFC (%)')
    plt.ylabel('Temperature (°C)')
    plt.title(title)
    plt.legend()
    if SAVE_FIGS: plt.savefig(f"{OUT_DIR}/{prefix}_contour.png", dpi=300)
    plt.show()

# Use global df for plotting the experimental run overlay
contour_and_surface(comp_F1, comp_F2, comp_pred, 'RSM Predicted - Compressive Strength', 'compressive')
contour_and_surface(flex_F1, flex_F2, flex_pred, 'RSM Predicted - Flexural Strength', 'flexural')

# -------------------------
# 7) Multi-response Optimization using Desirability (simple approach)
# -------------------------
# We predict compressive and flexural on the SAME grid (must be same axes)
# For safety, construct a common grid covering both datasets' union
union_df = pd.concat([comp_df[['BFC (%)','Temperature (°C)']], flex_df[['BFC (%)','Temperature (°C)']]])
union_min_f1, union_max_f1 = union_df['BFC (%)'].min(), union_df['BFC (%)'].max()
union_min_f2, union_max_f2 = union_df['Temperature (°C)'].min(), union_df['Temperature (°C)'].max()
F1u = np.linspace(max(0, union_min_f1 - 0.1*(union_max_f1-union_min_f1)), union_max_f1 + 0.1*(union_max_f1-union_min_f1), 101)
F2u = np.linspace(max(0, union_min_f2 - 0.1*(union_max_f2-union_min_f2)), union_max_f2 + 0.1*(union_max_f2-union_min_f2), 101)
F1uG, F2uG = np.meshgrid(F1u, F2u)
DFu = pd.DataFrame({
    'const': 1.0,
    'BFC (%)': F1uG.ravel(),
    'Temperature (°C)': F2uG.ravel(),
    'BFC (%)*Temperature (°C)': (F1uG*F2uG).ravel(),
    'BFC (%)^2': (F1uG**2).ravel(),
    'Temperature (°C)^2': (F2uG**2).ravel()
})
# Predict both responses (if model available)
pred_comp_on_union = comp_model.predict(DFu) if len(comp_df)>0 else pd.Series(np.nan, index=DFu.index)
pred_flex_on_union = flex_model.predict(DFu) if len(flex_df)>0 else pd.Series(np.nan, index=DFu.index)

# Scale predictions to [0,1] desirability (higher is better)
scaler = MinMaxScaler()
scaled_comp = scaler.fit_transform(pred_comp_on_union.values.reshape(-1,1)).ravel()
scaled_flex = scaler.fit_transform(pred_flex_on_union.values.reshape(-1,1)).ravel()

# Compute overall desirability = geometric mean (give equal weight). add small epsilon to avoid zeros
epsilon = 1e-8
desirability = np.sqrt((scaled_comp + epsilon) * (scaled_flex + epsilon))
desir_mat = desirability.reshape(F1uG.shape)
idx = np.unravel_index(np.nanargmax(desir_mat), desir_mat.shape)
opt_bfc_multi = F1uG[idx]
opt_temp_multi = F2uG[idx]
opt_comp_at_multi = pred_comp_on_union.values.reshape(F1uG.shape)[idx]
opt_flex_at_multi = pred_flex_on_union.values.reshape(F1uG.shape)[idx]
opt_desir = desir_mat[idx]

print(f"\nMulti-response optimum (desirability): BFC = {opt_bfc_multi:.4g} %, Temp = {opt_temp_multi:.3f} °C")
print(f" Predicted Compressive = {opt_comp_at_multi:.4f} MPa, Flexural = {opt_flex_at_multi:.4f} MPa, Desirability = {opt_desir:.4f}")

# Save grid predictions and desirability
grid_out = pd.DataFrame({
    'BFC (%)': DFu['BFC (%)'],
    'Temperature (°C)': DFu['Temperature (°C)'],
    'Pred_Compressive': pred_comp_on_union,
    'Pred_Flexural': pred_flex_on_union,
    'Desirability': desirability
})
grid_out.to_csv(f"{OUT_DIR}/prediction_grid_and_desirability.csv", index=False)
print(f"\nSaved prediction grid to {OUT_DIR}/prediction_grid_and_desirability.csv")

# -------------------------
# 8) Save regression coefficients and summaries
# -------------------------
def save_model_summary(model, name):
    coef_df = pd.DataFrame({'term': model.params.index, 'coef': model.params.values, 'pvalue': model.pvalues.values})
    coef_df.to_csv(f"{OUT_DIR}/{name}_model_coefs.csv", index=False)
    return coef_df

if len(comp_df)>0:
    save_model_summary(comp_model, 'compressive')
if len(flex_df)>0:
    save_model_summary(flex_model, 'flexural')

print("\nAll done. Outputs and figures are in the folder:", OUT_DIR)
```

```python
# =========================================
# RSM + CCD Analysis — Optimal Fiber % and Temperature
# =========================================
# Requirements:
# pip install pandas numpy scipy matplotlib seaborn statsmodels scikit-learn pyDOE2 openpyxl

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from sklearn.preprocessing import MinMaxScaler
from pyDOE2 import ccdesign
import os

# -------------------------
# USER CONFIG
# -------------------------
DATA_PATH = "dataset2.xlsx"   # <-- Your dataset (if available)
SHEET_NAME = 0
SAVE_FIGS = True
OUT_DIR = "rsm_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# ============================================================
# 0) OPTIONAL: GENERATE A CCD DESIGN (for verification or new run)
# ============================================================
# Factor setup
BFC_min, BFC_max = 0.0, 1.1         # Fiber %
Temp_min, Temp_max = 25, 500        # Temperature °C
k = 2  # number of factors

# Generate Central Composite Design (CCD)
ccd_matrix = ccdesign(k, center=(4,4), face='ccc')  # central composite circumscribed
ccd_df = pd.DataFrame(ccd_matrix, columns=['A_coded', 'B_coded'])

# Scale coded levels (-α to +α) to actual values
ccd_df['BFC (%)'] = (BFC_max - BFC_min)/2 * ccd_df['A_coded'] + (BFC_max + BFC_min)/2
ccd_df['Temperature (°C)'] = (Temp_max - Temp_min)/2 * ccd_df['B_coded'] + (Temp_max + Temp_min)/2
ccd_df['Run'] = range(1, len(ccd_df)+1)

print("\n=== Generated CCD Matrix (Preview) ===")
print(ccd_df.round(3).head(16))

# Save CCD layout for reference or experiment planning
ccd_df.to_excel(f"{OUT_DIR}/CCD_Design_Matrix.xlsx", index=False)
print(f"\nCCD matrix saved to: {OUT_DIR}/CCD_Design_Matrix.xlsx")

# ============================================================
# 1) LOAD DATA (if actual experimental results exist)
# ============================================================
if os.path.exists(DATA_PATH):
    df = pd.read_excel(DATA_PATH, sheet_name=SHEET_NAME)
    df.columns = df.columns.str.strip()
else:
    print("\n⚠️ No dataset found yet. Using CCD layout as template.")
    df = ccd_df.copy()

required_cols = ['BFC (%)', 'Temperature (°C)', 'Strength (MPa)']
# If Strength missing (i.e., no experiment data yet), fill with NaN
for col in required_cols:
    if col not in df.columns:
        df[col] = np.nan

# Add default dummy fields if not in data
for add_col in ['ID','FORCE (kN)','Duration (min)','Batch type']:
    if add_col not in df.columns:
        df[add_col] = np.nan

print("\nLoaded dataset with rows:", len(df))
print("Columns:", df.columns.tolist())

# Clean numeric fields
for c in ['FORCE (kN)', 'Strength (MPa)', 'BFC (%)', 'Temperature (°C)', 'Duration (min)']:
    df[c] = pd.to_numeric(df[c], errors='coerce')

# -------------------------
# 2) Split by Test Type
# -------------------------
if 'Batch type' in df.columns:
    df['TestType'] = df['Batch type'].apply(lambda x: 'Compressive' if isinstance(x, str) and 'COMPRESS' in x.upper()
                                            else ('Flexural' if isinstance(x, str) and 'FLEX' in x.upper() else 'Other'))
else:
    df['TestType'] = 'Compressive'  # default if missing

print(df['TestType'].value_counts())

comp_df = df[df['TestType']=='Compressive'].dropna(subset=['BFC (%)','Temperature (°C)']).copy()
flex_df = df[df['TestType']=='Flexural'].dropna(subset=['BFC (%)','Temperature (°C)']).copy()
print("\nCompressive rows:", len(comp_df), " | Flexural rows:", len(flex_df))

# -------------------------
# Helper: build quadratic design matrix
# -------------------------
def build_quadratic_X(df_in, factor1='BFC (%)', factor2='Temperature (°C)'):
    X = pd.DataFrame({
        'const': 1.0,
        factor1: df_in[factor1],
        factor2: df_in[factor2],
        f'{factor1}*{factor2}': df_in[factor1]*df_in[factor2],
        f'{factor1}^2': df_in[factor1]**2,
        f'{factor2}^2': df_in[factor2]**2
    })
    return X

# -------------------------
# 3) Fit quadratic RSM (OLS)
# -------------------------
def fit_rsm(df_in, response_col='Strength (MPa)', name='Compressive'):
    print(f"\n--- RSM for {name} (n={len(df_in)}) ---")
    X = build_quadratic_X(df_in)
    y = df_in[response_col]
    model = sm.OLS(y, X, missing='drop').fit()
    print(model.summary())
    try:
        anova_res = anova_lm(model, typ=2)
        print("\nANOVA (Type II):")
        print(anova_res)
    except Exception as e:
        print("ANOVA error:", e)
    return model, X, y

comp_model, comp_X, comp_y = fit_rsm(comp_df, name='Compressive')
flex_model, flex_X, flex_y = fit_rsm(flex_df, name='Flexural')

# -------------------------
# 4) Plot Diagnostics
# -------------------------
def diagnostics_plots(model, X, y, tag):
    resid = model.resid
    fitted = model.fittedvalues

    plt.figure(figsize=(6,4))
    plt.scatter(fitted, resid)
    plt.axhline(0, color='k', linestyle='--', linewidth=0.6)
    plt.xlabel('Fitted')
    plt.ylabel('Residuals')
    plt.title(f'{tag}: Residuals vs Fitted')
    if SAVE_FIGS: plt.savefig(f"{OUT_DIR}/{tag}_resid_vs_fitted.png", dpi=300)
    plt.show()

    sm.qqplot(resid, line='45', fit=True)
    plt.title(f'{tag}: QQ Plot of Residuals')
    if SAVE_FIGS: plt.savefig(f"{OUT_DIR}/{tag}_qq.png", dpi=300)
    plt.show()

if len(comp_df)>0: diagnostics_plots(comp_model, comp_X, comp_y, 'Compressive')
if len(flex_df)>0: diagnostics_plots(flex_model, flex_X, flex_y, 'Flexural')

# -------------------------
# 5) Response Surface Optimization Grid
# -------------------------
def prediction_grid_and_optimize(model, df_in, f1='BFC (%)', f2='Temperature (°C)', grid_n=101):
    f1_min, f1_max = df_in[f1].min(), df_in[f1].max()
    f2_min, f2_max = df_in[f2].min(), df_in[f2].max()
    f1_vals = np.linspace(f1_min, f1_max, grid_n)
    f2_vals = np.linspace(f2_min, f2_max, grid_n)
    F1, F2 = np.meshgrid(f1_vals, f2_vals)
    DF = pd.DataFrame({
        'const': 1.0,
        f1: F1.ravel(),
        f2: F2.ravel(),
        f'{f1}*{f2}': (F1*F2).ravel(),
        f'{f1}^2': (F1**2).ravel(),
        f'{f2}^2': (F2**2).ravel()
    })
    preds = model.predict(DF)
    preds_mat = preds.values.reshape(F1.shape)
    idx = np.unravel_index(np.nanargmax(preds_mat), preds_mat.shape)
    opt_f1 = F1[idx]
    opt_f2 = F2[idx]
    opt_pred = preds_mat[idx]
    return (F1, F2, preds_mat, opt_f1, opt_f2, opt_pred)

if len(comp_df)>0:
    comp_F1, comp_F2, comp_pred, comp_opt_f1, comp_opt_f2, comp_opt_pred = prediction_grid_and_optimize(comp_model, comp_df)
    print(f"\nCompressive Optimum: BFC={comp_opt_f1:.3f} %, Temp={comp_opt_f2:.1f} °C -> Strength={comp_opt_pred:.3f} MPa")

if len(flex_df)>0:
    flex_F1, flex_F2, flex_pred, flex_opt_f1, flex_opt_f2, flex_opt_pred = prediction_grid_and_optimize(flex_model, flex_df)
    print(f"Flexural Optimum: BFC={flex_opt_f1:.3f} %, Temp={flex_opt_f2:.1f} °C -> Strength={flex_opt_pred:.3f} MPa")



# =========================================
# 5B) 3D SURFACE & CONTOUR PLOTS (RSM Visualization)
# =========================================
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def rsm_surface_and_contour(F1, F2, pred_mat, f1_label, f2_label, title, prefix, 
                            exp_df=None, save=True):
    """Creates both 3D surface and 2D contour plots for RSM models."""
    
    # --- 3D Surface Plot ---
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(F1, F2, pred_mat, cmap=cm.viridis, alpha=0.85, linewidth=0, antialiased=True)
    ax.set_xlabel(f1_label)
    ax.set_ylabel(f2_label)
    ax.set_zlabel('Predicted Strength (MPa)')
    ax.set_title(f"{title} - 3D Surface")
    fig.colorbar(surf, shrink=0.5, aspect=10, pad=0.1, label='Strength (MPa)')
    if exp_df is not None:
        ax.scatter(exp_df[f1_label], exp_df[f2_label], exp_df['Strength (MPa)'], 
                   color='red', s=40, label='Experimental')
        ax.legend()
    if save:
        plt.savefig(f"{OUT_DIR}/{prefix}_3D_surface.png", dpi=300, bbox_inches='tight')
    plt.show()

    # --- 2D Contour Plot ---
    plt.figure(figsize=(8,6))
    cs = plt.contourf(F1, F2, pred_mat, cmap='viridis', levels=30)
    plt.colorbar(cs, label='Predicted Strength (MPa)')
    plt.xlabel(f1_label)
    plt.ylabel(f2_label)
    plt.title(f"{title} - Contour Plot")
    if exp_df is not None:
        plt.scatter(exp_df[f1_label], exp_df[f2_label], c='white', s=40, edgecolor='k', label='Experimental')
        plt.legend()
    if save:
        plt.savefig(f"{OUT_DIR}/{prefix}_contour.png", dpi=300, bbox_inches='tight')
    plt.show()

# ---- Generate RSM plots for each response ----
if len(comp_df) > 0:
    rsm_surface_and_contour(comp_F1, comp_F2, comp_pred, 
                            'BFC (%)', 'Temperature (°C)',
                            'Compressive Strength (RSM Surface)', 
                            'compressive', exp_df=comp_df)

if len(flex_df) > 0:
    rsm_surface_and_contour(flex_F1, flex_F2, flex_pred, 
                            'BFC (%)', 'Temperature (°C)',
                            'Flexural Strength (RSM Surface)', 
                            'flexural', exp_df=flex_df)

# -------------------------
# 6) Multi-response Desirability Optimization
# -------------------------
if len(comp_df)>0 and len(flex_df)>0:
    union_df = pd.concat([comp_df[['BFC (%)','Temperature (°C)']], flex_df[['BFC (%)','Temperature (°C)']]])
    F1u = np.linspace(union_df['BFC (%)'].min(), union_df['BFC (%)'].max(), 101)
    F2u = np.linspace(union_df['Temperature (°C)'].min(), union_df['Temperature (°C)'].max(), 101)
    F1uG, F2uG = np.meshgrid(F1u, F2u)
    DFu = pd.DataFrame({
        'const': 1.0,
        'BFC (%)': F1uG.ravel(),
        'Temperature (°C)': F2uG.ravel(),
        'BFC (%)*Temperature (°C)': (F1uG*F2uG).ravel(),
        'BFC (%)^2': (F1uG**2).ravel(),
        'Temperature (°C)^2': (F2uG**2).ravel()
    })
    pred_comp = comp_model.predict(DFu)
    pred_flex = flex_model.predict(DFu)
    scaler = MinMaxScaler()
    d1 = scaler.fit_transform(pred_comp.values.reshape(-1,1)).ravel()
    d2 = scaler.fit_transform(pred_flex.values.reshape(-1,1)).ravel()
    desir = np.sqrt(d1*d2 + 1e-8)
    des_mat = desir.reshape(F1uG.shape)
    idx = np.unravel_index(np.nanargmax(des_mat), des_mat.shape)
    print(f"\nMulti-response Optimum (Desirability): BFC={F1uG[idx]:.3f} %, Temp={F2uG[idx]:.1f} °C, Desir={des_mat[idx]:.3f}")
```
